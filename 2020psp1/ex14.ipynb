## このテキストセルを使って、以下のコマンドセルの全体の概要を説明してください。

個別のコマンドについての解説はコマンドセル中のコマンドの横に #でコメントを書くようにして、テキストセルでは複数のセルの一連の動作概要を書くと見やすくなると思います。

```
import pandas as pd
import numpy as np
```

```
df = pd.read_excel('data_from_wget.xls')
df
```

```
df = df.rename(columns={'（Ｐ６）　我が国の人口ピラミッド－平成28年10月1日現在－':'age','Unnamed: 1':'male','Unnamed: 2':'female'})
df
```

```
df = df[1:102]
df
```

```
age=np.array(df['age']).astype(str)
age[100]='100+'
print(age)
```

```
male=np.array(df['male'])
print(male)
```

```
female=np.array(df['female'])
print(female)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10)) # 大きめのグラフを指定して見やすくする
plt.title('Japanese population') # グラフのタイトル
plt.xticks(np.arange(0,101,5)) # デフォルトだとx軸の文字が重なるので5歳おきに表示
plt.plot(age,male,label='male') # maleのグラフ
plt.plot(age,female,label='female') # femaleのグラフ
plt.legend() # グラフの凡例のことをlegendという
plt.savefig('japanese_population.png') # ディスクに japanese_population.png として保存（ローカルのコンピュータに取り込むには更にダウンロードして）
```
